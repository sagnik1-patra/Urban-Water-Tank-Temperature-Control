```python
import os
import json
import yaml
import joblib
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Urban Water Tank Temperature Control"
DATA_PATH = os.path.join(
    BASE_DIR,
    "results_all_improvements_city_to_tourism.xlsx"
)

os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_excel(DATA_PATH)
df.columns = df.columns.str.lower()
df = df.select_dtypes(include=[np.number]).dropna()

X_full = df.iloc[:, :-1]
y = df.iloc[:, -1]

# ============================================================
#  AIS FEATURE SELECTION
# ============================================================

def ais_feature_selection(X, y, population=10, generations=5):
    n_features = X.shape[1]
    best_mask = None
    best_rmse = float("inf")

    for _ in range(generations):
        for _ in range(population):
            mask = np.random.randint(0, 2, n_features)
            if mask.sum() == 0:
                continue

            X_sel = X.iloc[:, mask == 1]
            X_tr, X_te, y_tr, y_te = train_test_split(
                X_sel, y, test_size=0.2, random_state=42
            )

            scaler = StandardScaler()
            X_tr = scaler.fit_transform(X_tr)
            X_te = scaler.transform(X_te)

            model = Sequential([
                Dense(16, activation="relu", input_shape=(X_tr.shape[1],)),
                Dense(1)
            ])
            model.compile(optimizer="adam", loss="mse")
            model.fit(X_tr, y_tr, epochs=12, verbose=0)

            pred = model.predict(X_te).flatten()
            rmse = np.sqrt(mean_squared_error(y_te, pred))

            if rmse < best_rmse:
                best_rmse = rmse
                best_mask = mask

    return best_mask

selected_mask = ais_feature_selection(X_full, y)
X_selected = X_full.iloc[:, selected_mask == 1]

# ============================================================
#  PSO FEATURE WEIGHT OPTIMIZATION
# ============================================================

def pso_optimize_weights(X, y, particles=12, iterations=15):
    dim = X.shape[1]
    swarm = np.random.rand(particles, dim)
    velocity = np.zeros((particles, dim))
    pbest = swarm.copy()
    pbest_score = np.full(particles, np.inf)
    gbest = None
    gbest_score = np.inf

    for _ in range(iterations):
        for i in range(particles):
            X_weighted = X * swarm[i]

            X_tr, X_te, y_tr, y_te = train_test_split(
                X_weighted, y, test_size=0.2, random_state=42
            )

            scaler = StandardScaler()
            X_tr = scaler.fit_transform(X_tr)
            X_te = scaler.transform(X_te)

            model = Sequential([
                Dense(16, activation="relu", input_shape=(X_tr.shape[1],)),
                Dense(1)
            ])
            model.compile(optimizer="adam", loss="mse")
            model.fit(X_tr, y_tr, epochs=10, verbose=0)

            pred = model.predict(X_te).flatten()
            rmse = np.sqrt(mean_squared_error(y_te, pred))

            if rmse < pbest_score[i]:
                pbest_score[i] = rmse
                pbest[i] = swarm[i].copy()

            if rmse < gbest_score:
                gbest_score = rmse
                gbest = swarm[i].copy()

        velocity = (
            0.5 * velocity
            + 1.5 * np.random.rand() * (pbest - swarm)
            + 1.5 * np.random.rand() * (gbest - swarm)
        )
        swarm += velocity

    return gbest

best_weights = pso_optimize_weights(X_selected, y)
X_final = X_selected * best_weights

# ============================================================
# FINAL TRAINING
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

joblib.dump(scaler, os.path.join(BASE_DIR, "pis_scaler.pkl"))

model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_s.shape[1],)),
    Dense(32, activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train_s, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
    verbose=1
)

train_pred = model.predict(X_train_s).flatten()
test_pred = model.predict(X_test_s).flatten()

# ============================================================
# METRICS
# ============================================================

metrics = {
    "train_rmse": float(np.sqrt(mean_squared_error(y_train, train_pred))),
    "test_rmse": float(np.sqrt(mean_squared_error(y_test, test_pred))),
    "train_r2": float(r2_score(y_train, train_pred)),
    "test_r2": float(r2_score(y_test, test_pred)),
    "selected_features": list(X_selected.columns),
    "pso_feature_weights": best_weights.tolist()
}

# ============================================================
# VISUALIZATION (SHOW + SAVE)
# ============================================================

plt.figure()
plt.plot(history.history["loss"], label="Train")
plt.plot(history.history["val_loss"], label="Validation")
plt.legend()
plt.title("PIS Accuracy Loss Curve")
plt.show()
plt.savefig(os.path.join(BASE_DIR, "pis_accuracy_loss.png"))

plt.figure()
plt.scatter(y_test, test_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("PIS Actual vs Predicted")
plt.show()
plt.savefig(os.path.join(BASE_DIR, "pis_actual_vs_predicted.png"))

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("PIS Feature Heatmap")
plt.show()
plt.savefig(os.path.join(BASE_DIR, "pis_heatmap.png"))

plt.figure()
plt.plot(test_pred, label="Predicted")
plt.plot(y_test.values, label="Actual")
plt.legend()
plt.title("PIS Prediction Trend")
plt.show()
plt.savefig(os.path.join(BASE_DIR, "pis_prediction_trend.png"))

# ============================================================
# SAVE FILES
# ============================================================

model.save(os.path.join(BASE_DIR, "pis_ann_model.h5"))

pd.DataFrame({
    "actual": y_test.values,
    "predicted": test_pred
}).to_csv(
    os.path.join(BASE_DIR, "pis_results.csv"),
    index=False
)

with open(os.path.join(BASE_DIR, "pis_predictions.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(BASE_DIR, "pis_model_config.yaml"), "w") as f:
    yaml.dump(metrics, f)

print("\n PIS (PSO + AIS) PIPELINE COMPLETED SUCCESSFULLY")
print(" All outputs saved with 'pis_' prefix")
```
